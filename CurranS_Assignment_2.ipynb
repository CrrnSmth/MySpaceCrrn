```python
import gradio as gr
from transformers import pipeline

# Load translation models for different languages
translator_models = {
    "French": pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr"),
    "Spanish": pipeline("translation_en_to_es", model="Helsinki-NLP/opus-mt-en-es"),
    "Hindi": pipeline("translation_en_to_hi", model="Helsinki-NLP/opus-mt-en-hi"),
    "Chinese": pipeline("translation_en_to_zh", model="Helsinki-NLP/opus-mt-en-zh"),
    # "Portuguese": pipeline("translation_en_to_pt", model="Helsinki-NLP/opus-mt-en-pt"),  # This model seems to be unavailable or private. Removing it or finding an alternative is recommended.
    "Russian": pipeline("translation_en_to_ru", model="Helsinki-NLP/opus-mt-en-ru"),
    "Arabic": pipeline("translation_en_to_ar", model="Helsinki-NLP/opus-mt-en-ar"),
}

# Translation function
def translate_text(text, language):
    if language in translator_models:
        translated = translator_models[language](text)[0]['translation_text']
    else:
        translated = "Unsupported language selected."  # This will handle the case where "Portuguese" is selected now.
    return translated

# Gradio Interface
with gr.Blocks() as demo:
    gr.Markdown("### English to Multiple Languages Translator")

    with gr.Row():
        input_text = gr.Textbox(label="Enter English Text", placeholder="Type here...")
        # Remove "Portuguese" from the dropdown options:
        language = gr.Dropdown(
            ["French", "Spanish", "Hindi", "Chinese", "Russian", "Arabic"],
            value="French",
            label="Select Language"
        )

    output_text = gr.Textbox(label="Translated Text")
    translate_button = gr.Button("Translate")

    translate_button.click(translate_text, inputs=[input_text, language], outputs=output_text)

# Run the app
demo.launch()
```
